```python
import pandas as pd
pd.set_option('display.max_columns',None)
system = pd.read_csv("Sim_data_repository_5/simdata_steer_fault_bias1006.csv")
```

```python
system.head()
```

```python
sys = system.fillna(5, axis = 1)
sys
```

```python
bins = [0,0.3,0.6,1,10]
labels = [0, 1, 2,"?"]
list = sys.columns
learningData = pd.DataFrame()
for i in list :
    learningData[i] = pd.cut(sys[i], bins, include_lowest=True, labels=labels)
#     type = sys.loc[:, [i]].dtypes == 'float'
#     if type.bool() : 
#         learningData[i] = pd.cut(sys[i] , bins, include_lowest=True, labels=labels)
#     else :
#         learningData[i] = sys[i]
```

```python
learningData
```

```python
for i in list :
    print(i)
```

```python
learningData = learningData.rename(columns={'cur_vel_1':'cur_vel_x_1', 'cur_vel_1.1':'cur_vel_y_1','position_front_car_1':'position_front_car_y_1',
                                           'position_around2_1':'position_around2_x_1','position_around2_1.1':'position_around2_y_1',
                                           'speed_around3_t_0':'speed_around3_0'})
learningData = learningData.drop(['Time','fault_location','fault_type','fault_start_time',
                                  'number_of_cars','fault_duration','fault_value'],
                                axis=1)
list = learningData.columns
for i in list :
    print(i)
```

```python
learner = gum.BNLearner('learningData_1.csv',dbn, ["?"])
print(f"Missing values in {learningData} : {learner.hasMissingValues()}")
```

```python
learner.useEM(1e-2)
learner.useAprioriSmoothing()
print(learner)
```

```python
import matplotlib.pyplot as plt
import numpy as np


try:  
    learner.learnParameters(dbn.dag())
    learner.currentTime()

    plt.plot(np.arange(1,1+learner.nbrIterations()),learner.history())
    plt.xticks(np.arange(1, 1+learner.nbrIterations(), step=2))
    plt.title("Error during EM iterations");
    
except gum.MissingValueInDatabase:
  print("Learning is not possible without EM if there are some missing values.")
```
